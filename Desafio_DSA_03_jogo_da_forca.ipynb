```python
#Jupyter Notebook - Lab 03 DSA Jogo da forca

import random

#board game, cria uma lista de elementos com indice de 0 a 6 chamado board

board = ['''
>Hangman<
+---+
|   |
    |
    |
    |
    |
=========''', '''
+---+
|   |
O   |
    |
    |
    |
=========''', '''
+---+
|   |
O   |
|   |
    |
    |
=========''', '''
-+--+
 |  |
 O  |
/|  |
    |
    |
=========''', '''
-+--+
 |  |
 O  |
/|\ |
    |
    |
=========''', '''
-+--+
 |  |
 O  |
/|\ |
/   |
    |
=========''', '''
-+--+
 |  |
 O  |
/|\ |
/ \ |
    |
=========''']


class Hangman:
  # metodo construtor, inicializa o objeto, self e word, cria listas vazias
  def __init__(self, word):
    self.word = word
    self.missed_letters = []
    self.guessed_letters = []
  
  # metodo para adivinhação da letra
  def guess(self, letter):
    if letter in self.word and letter not in self.guessed_letters:
      self.guessed_letters.append(letter)
    elif letter not in self.word and letter not in self.missed_letters:
      self.missed_letters.append(letter)
    else:
      return False
    return True

  # metodo para nao mostrar a palavra no board
  def hide_word(self):
      rtn = " "
      for letter in self.word:
        if letter not in self.guessed_letters:
          rtn += "_"
        else:
          rtn += letter
      return rtn
  
  # metodo para checar se o jogador venceu
  def hangman_won(self):
    if "_" not in self.hide_word():
      return True
    return False
 
 # metodo par checar se o jogo acabou
  def hangman_over(self):
    return self.hangman_won() or (len(self.missed_letters) == 6)

  # checar o status do game e imprimir o board na tela
  def print_game_status(self):
    print(board[len(self.missed_letters)])
    print("\nPalavra : " + self.hide_word())
    print("\nLetras erradas : ", )
    for letter in self.missed_letters:
      print(letter,)
    print()
    print("\nLetras corretas : ",)
    for letter in self.guessed_letters:
      print(letter,)
    print()

# Metodos anteriores fazem parte da classe hangman, comprovado pela identacao

# agora eu teria que definir um metodo para a palavra que o jogo vai usar
# eu poderia utilizar o banco de palavras passado pela DSA
# mas como eu quero que o jogo possa ser inicializado em condicoes adversas
# dessa forma vou inicializar o jogo com uma palavra fornecida por um jogador p
# o outro jogador tentar acertar

def ask_word():
  palavra = input(str("\n Escolha uma palavra para jogar : "))
  return palavra

# metodo main e execucao do programa
def main():
  #objeto
  game = Hangman(ask_word())

  #enquanto o jogo nao tiver terminado ele continua hahaha
  while not game.hangman_over():
    game.print_game_status()
    usr_input = input("\nDigite uma letra para jogar : ")
    game.guess(usr_input)

  game.print_game_status()

  #personalizar uma mensagem para cada status de jogo
  if game.hangman_won():
    print("\nParabens voce venceu!")
  else:
    print("\nVoce perdeu!")
    print("\nA palavra era:")
    print(game.word)
  
  print("\nFoi bom jogar com vc")

if __name__ == "__main__":
  main()
```
